```python
### PCA Example 1
```

```python
nsample = 1000

v1 = np.random.normal(0, 2, size=nsample)
v2 = 2*np.random.weibull(a=0.9, size=nsample)
v3 = np.random.weibull(a=2, size=nsample)
e4 = np.random.normal(0,3)
v4 = -0.3*v1 + v2 + 0.5*v3 + e4


X = []
for i in range(5): # 0-4
    X.append(v1 + np.random.normal(0,1))
for i in range(5): # 5-9
    X.append(v2 + np.random.normal(0,3))
for i in range(5): # 10-14
    X.append(v3 + np.random.normal(0,3) - 0.3*v2)
for i in range(2): # 15, 16
    X.append(v4 + np.random.normal(0,3))
X = np.vstack(X).T
```

```python
plot_distribution(X[:,[0,5,10,15,16]])
```

```python
np.save('./pca_ex1.npy', X)
```

```python
### PCA Example 2
```

```python
nsample = 1000

v1 = np.random.normal(0, 100, size=nsample)
v2 = 5*np.random.weibull(a=0.8, size=nsample)
v3 = np.random.gamma(2, 2, size=nsample)
e4 = np.random.normal(0,3)
v4 = v2 + 0.5*v3


X = []
for i in range(5): # 0-4
    X.append(v1 + np.random.normal(0,i*2) + i*50)
for i in range(5): # 5-9
    X.append((1+i)*v2 + np.random.normal(0,1))
for i in range(5): # 10-14
    X.append((1+i)*v3 + np.random.normal(0,1))
for i in range(2): # 15, 16
    X.append(v4 + np.random.normal(0,3*i))
X = np.vstack(X).T
```

```python
plot_distribution(X[:,[0,5,10,15,16]])
```

```python
np.save('./pca_ex2.npy', X)
```

```python
### CCA Example 1
```

```python
nsample = 5000

v1 = np.random.normal(0, 5, size=nsample)
v2 = np.random.lognormal(1, 0.5, size=nsample)
v3 = np.random.gamma(2, 2, size=nsample)
v4 = np.random.logistic(20, scale=2, size=nsample)

V = np.vstack((v1,v2,v3,v4)).T

X = []
for i in range(5): # 0-4
    X.append(shuffle(v3 + np.random.normal(0,10), 10))
for i in range(4): # 5-9
    X.append(shuffle(v2 + np.random.normal(0,3), 10))
for i in range(3): # 10-19
    X.append(shuffle(-0.5*v2 + v3 + np.random.normal(0,5), 10))
X = np.vstack(X).T
plot_distribution(X, ncols=7)
```

```python
Y = []
for i in range(5): # 0-4
    Y.append(shuffle(v1 + 5*v3 + np.random.normal(0,1), 10))
for i in range(5): # 5-9
    Y.append(0.5*v4 + v2 + np.random.normal(0,5))
Y = np.vstack(Y).T
plot_distribution(Y, ncols=7)
```

```python
np.savez('./CCA_ex1.npz', X=X, Y=Y, V=V)
```

```python
### CCA Example 2
```

```python
nsample = 5000

v1 = np.random.normal(0, 5, size=nsample)
v2 = np.random.lognormal(1, 0.5, size=nsample)
v3 = np.random.gamma(2, 2, size=nsample)
v4 = np.random.logistic(20, scale=2, size=nsample)

V = np.vstack((v1,v2,v3,v4)).T

plot_distribution(V, ncols=4)
```

```python
X = []
for i in range(5): # 0-4
    X.append(shuffle(v3 + np.random.normal(0,10), 10))
for i in range(4): # 5-9
    X.append(shuffle(v2 + np.random.normal(0,3), 10))
for i in range(3): # 10-19
    X.append(shuffle(-0.5*v2 + v3 + np.random.normal(0,5), 10))
X = np.vstack(X).T
plot_distribution(X, ncols=7)
```

```python
Y = []
for i in range(5): # 0-4
    Y.append(shuffle(v1 + 0.5*v3 + np.random.normal(0,1), 20))
for i in range(5): # 5-9
    Y.append(shuffle(v4 + 0.3*v2 + np.random.normal(0,5), 20))
Y = np.vstack(Y).T
plot_distribution(Y, ncols=7)
```

```python
np.savez('./CCA_ex2.npz', X=X, Y=Y, V=V)
```
