## PCA Example 1

```python
X = np.load('./pca_ex1.npy')
```

```python
plot_distribution(X[:,[0,5,10,15]])
```

```python
plot_distribution(X, ncols=7)
```

```python
PCs, vari_explain, s = pca_wrap(X, method='vanilla')
```

```python
plot_distribution(PCs, ncols=7, title_list=['{:0.2e}'.format(v) for v in vari_explain])
```

```python
print(vari_explain.cumsum())
```

```python
PCs, vari_explain, s = pca_wrap(X, method='sPCA', reg=1, ridge_reg=0.1)
```

```python
plot_distribution(PCs, ncols=7, title_list=['{:0.2e}'.format(v) for v in vari_explain])
```

```python
print(vari_explain.cumsum())
```

```python
PCs, vari_explain, s = pca_wrap(X, method='rPCA', reg=0.01)
```

```python
plot_distribution(PCs, ncols=7, title_list=['{:0.2e}'.format(v) for v in vari_explain])
```

```python
print(vari_explain.cumsum())
```

## PCA Example 2

```python
X = np.load('./pca_ex2.npy')
```

```python
plot_distribution(X[:,[0,5,10,15]])
```

```python
plot_distribution(X, ncols=7)
```

```python
PCs, vari_explain, s = pca_wrap(X, method='vanilla')
plot_distribution(PCs, ncols=7, title_list=['{:0.2e}'.format(v) for v in vari_explain])
print(vari_explain.cumsum())
```

```python
PCs, vari_explain, s = pca_wrap(X, method='sPCA', reg=1, ridge_reg=0.1)
plot_distribution(PCs, ncols=7, title_list=['{:0.2e}'.format(v) for v in vari_explain])
print(vari_explain.cumsum())
```

```python
PCs, vari_explain, s = pca_wrap(X, method='rPCA', reg=0.005)
plot_distribution(PCs, ncols=7, title_list=['{:0.2e}'.format(v) for v in vari_explain])
print(vari_explain.cumsum())
```
