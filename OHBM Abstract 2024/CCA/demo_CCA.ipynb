```python
import numpy as np
import scipy
from utils import *

from sklearn.cross_decomposition import CCA
```

## CCA Example 1

```python
data = np.load('./CCA_ex1.npz')
X = data['X']
Y = data['Y']
V = data['V']
```

```python
Xpc, X_vari_explain, _ = pca_wrap(X, method='sPCA', reg=1, ridge_reg=0.1)
Ypc, Y_vari_explain, _ = pca_wrap(Y, method='sPCA', reg=1, ridge_reg=0.1)
```

```python
k = 7
fig, ax = pyplot.subplots(figsize=(4,4))
ax.plot(np.arange(k)+1, X_vari_explain[:k], 'r')
ax.plot(np.arange(k)+1, Y_vari_explain[:k], 'b')
pyplot.show()
```

```python
k = 7
cca = CCA(n_components=k)
cca.fit(Xpc[:,:k], Ypc[:,:k])
Xc, Yc = cca.transform(Xpc[:,:k], Ypc[:,:k])
```

```python
plot_distribution(V)
plot_distribution(Xc)
plot_distribution(Yc)
```

```python
n = X.shape[1]
m = V.shape[1]
corrcoef = np.corrcoef(np.hstack((X, V)).T)
corrcoef = corrcoef[0:n,:][:,n:].T
fig, ax = pyplot.subplots(figsize=(6,3))
sns.heatmap(corrcoef, annot=True, annot_kws={"fontsize":6}, ax=ax, cmap='Blues',
            xticklabels=[f"X-{i+1}" for i in range(n)],
            yticklabels=[f"V-{i+1}" for i in range(m)])
```

```python
corrcoef = np.corrcoef(np.hstack((Xc, V)).T)
corrcoef = corrcoef[0:k,:][:,k:].T
fig, ax = pyplot.subplots(figsize=(6,3))
sns.heatmap(corrcoef, annot=True, annot_kws={"fontsize":6}, ax=ax, cmap='Blues',
            xticklabels=[f"Xc-{i+1}" for i in range(k)],
            yticklabels=[f"V-{i+1}" for i in range(m)])
```

```python
n = Y.shape[1]
corrcoef = np.corrcoef(np.hstack((Y, V)).T)
corrcoef = corrcoef[0:n,:][:,n:].T
fig, ax = pyplot.subplots(figsize=(6,3))
sns.heatmap(corrcoef, annot=True, annot_kws={"fontsize":6}, ax=ax, cmap='Blues',
            xticklabels=[f"X-{i+1}" for i in range(n)],
            yticklabels=[f"V-{i+1}" for i in range(m)])
```

```python
corrcoef = np.corrcoef(np.hstack((Yc, V)).T)
corrcoef = corrcoef[0:k,:][:,k:].T
fig, ax = pyplot.subplots(figsize=(6,3))
sns.heatmap(corrcoef, annot=True, annot_kws={"fontsize":6}, ax=ax, cmap='Blues',
            xticklabels=[f"Yc-{i+1}" for i in range(k)],
            yticklabels=[f"V-{i+1}" for i in range(m)])
```

```python
corrcoef = np.corrcoef(np.hstack((Xc, Yc)).T)
corrcoef = corrcoef[0:k,:][:,k:].T
sns.heatmap(corrcoef, annot=True, annot_kws={"fontsize":6}, cmap='Blues',
            xticklabels=[f"Xc-{i+1}" for i in range(k)],
            yticklabels=[f"Yc-{i+1}" for i in range(k)])
```
